```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from astropy.table import Table, unique
#import astropy
#from astropy.io import fits

data = Table.read('zcat.deep2.dr4.uniq.fits')
deep=data.to_pandas()
```
